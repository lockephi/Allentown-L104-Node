````python
# Cell 1: System Setup & Disk Cleanup
import os
import shutil
import json
import subprocess
from datetime import datetime

print("=" * 60)
print("ğŸ§¹ DISK SPACE CLEANUP")
print("=" * 60)

# Check current disk usage
total, used, free = shutil.disk_usage("/")
print(f"\nğŸ“Š Before Cleanup:")
print(f"   Total: {total // (1024**3)} GB")
print(f"   Used:  {used // (1024**3)} GB")
print(f"   Free:  {free // (1024**3)} GB ({100*free/total:.1f}%)")

# Cleanup operations
cleanup_dirs = [
    os.path.expanduser("~/.cache/pip"),
    "/tmp",
]

bytes_freed = 0
for path in cleanup_dirs:
    if os.path.exists(path):
        try:
            size = sum(os.path.getsize(os.path.join(dp, f)) 
                      for dp, dn, filenames in os.walk(path) 
                      for f in filenames)
            bytes_freed += size
        except:
            pass

# Clean __pycache__ in workspace
workspace = "/workspaces/Allentown-L104-Node"
for root, dirs, files in os.walk(workspace):
    for d in dirs:
        if d == "__pycache__":
            try:
                shutil.rmtree(os.path.join(root, d))
            except:
                pass

# Clean .pyc files
for root, dirs, files in os.walk(workspace):
    for f in files:
        if f.endswith('.pyc'):
            try:
                os.remove(os.path.join(root, f))
            except:
                pass

print(f"\nâœ… Cleanup complete!")
print(f"   Estimated space recoverable: {bytes_freed // (1024**2)} MB")
````
